```python
# resizing the image using pillow library
from PIL import Image
```

```python
img = Image.open('dog.jpg')
img_resized = img.resize((200, 200))
```

```python
img_resized.save('dog_image_resized.jpg')
```

convert rgb images to Grayscale image using openCV

```python
# importing OpenCV library
import cv2
```

```python
img = cv2.imread('dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
from google.colab.patches import cv2_imshow
```

```python
# displaying the image
cv2_imshow(grayscale_image)
```

```python
# saving rhe grayscale image
cv2.imwrite('dog_graysale_image.jpg', grayscale_image)
```
